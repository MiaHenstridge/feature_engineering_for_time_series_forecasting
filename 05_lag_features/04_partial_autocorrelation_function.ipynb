
# Computing the partial autocorrelation function (PACF)
Feature Engineering for Time Series Forecasting

In this notebook we show how we can compute the PACF and PACF plots.

We will work with a monthly retail sales dataset (found here).

For instructions on how to download, prepare, and store the dataset, refer to notebook number 1, in the folder "01-Datasets" from this repo.

## Data Set Synopsis
The timeseries is collected between January 1992 and May 2016.

It consists of a single series of monthly values representing sales volumes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.nonparametric.smoothers_lowess import lowess
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import STL

sns.set_context('talk')
```

## Computing the PACF for a white noise timeseries

```python
num_timesteps = 300
np.random.seed(0)

y = np.random.normal(loc=0, scale=1, size=num_timesteps)
ts = pd.date_range(start='2000-01-01', periods=num_timesteps, freq='D')

df = pd.DataFrame(data={'y': y}, index=ts)
df.head()
```

```python
# plot timeseries
ax = df.plot(figsize=(10, 5))

ax.set_title('white noise')
ax.set_xlabel('time')
ax.set_ylabel('y')
```

```python
# create a copy to preserve the original dataframe for later use
df_ = df.copy()
```

```python
# plot pacf
fig, ax = plt.subplots(figsize=(10, 5))

plot_pacf(x=df_['y'],
         method='ywm',
         lags=50,
         ax=ax,
         alpha=0.05,
         auto_ylims=True)

ax.set_title('Partial autocorrelation of white noise')
ax.set_ylabel('Partial autocorrelation')
ax.set_xlabel('Lag')

plt.tight_layout()
```

* The partial autocorrelation is one at lag 0 as expected
* The partial autocorrelation at all other lags are not significant. As expected from white noise

We saw clearly that there is no correlation in the lag plots shown in notebook 3 on the ACF.

Now let's look at some timeseries with more structure and see what the PACF looks like.

## AR(1) process

```python
num_timesteps = 1000
np.random.seed(0)

phi = 0.9
ts = pd.date_range(start='2000-01-01', periods=num_timesteps, freq='D')

# generate time series
y = np.zeros(num_timesteps)
for t in range(1, num_timesteps):
    noise = np.random.normal()
    y[t] = phi * y[t-1] + noise

df = pd.DataFrame(data={'y': y}, index=ts)

df.head()
```

```python
# Plot the time series
ax = df.plot(figsize=(10, 5))
ax.set_title('AR1 process')
ax.set_xlabel('Time')
ax.set_ylabel('y')
```

```python
df_ = df.copy()
```

```python
# plot ACF & PACF
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# ACF
plot_acf(x=df_['y'],
        lags=36,
        ax=ax[0], 
        alpha=0.05,
        auto_ylims=True)

ax[0].set_title('Autocorrelation of AR(1) process')
ax[0].set_ylabel('Autocorrelation')
ax[0].set_xlabel('Lag')

# PACF
plot_pacf(x=df_['y'],
         lags=36,
         ax=ax[1],
         alpha=0.05,
         method='ywm',
         auto_ylims=True)

ax[1].set_title('Partial autocorrelation of AR(1) process')
ax[1].set_ylabel('Partial autocorrelation')
ax[1].set_xlabel('Lag')

plt.tight_layout()
```

The PACF is large for k = 1 and not significant for k > 2 as we wanted. Hence, the PACF can identify in this case that lag 1 is important

## AR(2) process

$$y_t = \phi_1y_{t-1} + \phi_2y_{t-2} + noise_t$$

where
$$noise_t ~ N(mean=0, std=1)$$

For the AR(1) process to be stationary, we required $|\phi| < 1$. To ensure that an AR(2) process remains stationary it can be shown the following constraints on $\phi_1$ and $\phi_2$ are required:
$$\phi_1 + \phi_2 < 1$$
$$-\phi_1 + \phi_2 < 1$$
$$\phi_2 > -1$$

```python
# generate time series
num_timesteps = 1000
np.random.seed(0)

phi_1 = 0.7
phi_2 = 0.2
ts = pd.date_range(start='2000-01-01', periods=num_timesteps, freq='D')

y = np.zeros(num_timesteps)
for t in range(2, num_timesteps):
    noise = np.random.normal()
    y[t] = phi_1 * y[t-1] + phi_2 *y[t-2] + noise

df = pd.DataFrame(data={'y': y}, index=ts)

df.head()
```

```python
# Plot the time series
ax = df.plot(figsize=(10, 5))
ax.set_title('AR2 process')
ax.set_xlabel('Time')
ax.set_ylabel('y')
```

```python
df_ = df.copy()
```

```python
# plot ACF & PACF
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# ACF
plot_acf(x=df_['y'],
        lags=36,
        ax=ax[0], 
        alpha=0.05,
        auto_ylims=True)

ax[0].set_title('Autocorrelation of AR(2) process')
ax[0].set_ylabel('Autocorrelation')
ax[0].set_xlabel('Lag')

# PACF
plot_pacf(x=df_['y'],
         lags=36,
         ax=ax[1],
         alpha=0.05,
         method='ywm',
         auto_ylims=True)

ax[1].set_title('Partial autocorrelation of AR(2) process')
ax[1].set_ylabel('Partial autocorrelation')
ax[1].set_xlabel('Lag')

plt.tight_layout()
```

* The PACF has significant lags at k=1 and k=2
* The PACF is larger for k=1 than k=2 because $|\phi_1| > |\phi_2|$
* This shows that the PACF can identify useful lags in this scenario.

## Time series with trend and seasonality

```python
# load retail sales datasets with artificially added outliers
df = pd.read_csv('../datasets/example_retail_sales.csv',
                parse_dates=['ds'],
                index_col=['ds'])
```

```python
# plot the time series
ax = df.plot(y='y', marker='.', figsize=(10, 5))

ax.set_title('retail sales')
ax.set_ylabel('Sales')
ax.set_xlabel('Time')
```

```python
df_ = df.copy()
```

```python
# plot ACF & PACF
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# ACF
plot_acf(x=df_['y'],
        lags=36,
        ax=ax[0], 
        alpha=0.05,
        auto_ylims=True)

ax[0].set_title('Autocorrelation of retail sales')
ax[0].set_ylabel('Autocorrelation')
ax[0].set_xlabel('Lag')

# PACF
plot_pacf(x=df_['y'],
         lags=36,
         ax=ax[1],
         alpha=0.05,
         method='ywm',
         auto_ylims=True)

ax[1].set_title('Partial autocorrelation of retail sales')
ax[1].set_ylabel('Partial autocorrelation')
ax[1].set_xlabel('Lag')

plt.tight_layout()
```

* The PACF assumes the time series is stationary
* This time series is clearly not stationary. The trend means that the mean changes in time.
* Despite that you still see a strong peak at k=12 and the smaller lags

Let's remove the trend and re-examine the PACF

```python
# compute trend with LOWESS
frac = 0.1
y = df_['y']
x = np.arange(0, len(y))

res = lowess(endog=y, exog=x, frac=frac)
df_['lowess'] = res[:, 1]

df_['y_detrend'] = df_['y'] - df_['lowess']
df_['y_detrend'].plot(figsize=(10, 5))
```

```python
# plot ACF & PACF
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# ACF
plot_acf(x=df_['y_detrend'],
        lags=36,
        ax=ax[0], 
        alpha=0.05,
        auto_ylims=True)

# ax[0].set_title('Autocorrelation of retail sales')
# ax[0].set_ylabel('Autocorrelation')
ax[0].set_xlabel('Lag')

# PACF
plot_pacf(x=df_['y_detrend'],
         lags=36,
         ax=ax[1],
         alpha=0.05,
         method='ywm',
         auto_ylims=True)

# ax[1].set_title('Partial autocorrelation of retail sales')
# ax[1].set_ylabel('Partial autocorrelation')
ax[1].set_xlabel('Lag')

plt.tight_layout()
```

The PACF shows a strong lag at 12 but not at multiples of 12 afterwards. This means that most of the predictive info from the seasonal lag is captured at lag 12 as expected!

In summary this would suggest that we could create features using a lag of 12 (from the seasonality) and low lags like 1, 2, 3, and 4. The other larger lags such as lag 10, 11, or 13 are more difficult to interpret. One reason for not including them is that you do not see peaks at multiples of 10, 11, or 13 in the ACF. Also from domain knowledge (this is retail sales) it is highly unlikely that lags beyond 12 months will be that relevant.

## PACF after detrending and deseasonalizing the original series


Now that we know the period of any seasonality we can remove the impact of the trend and seasonal components on the ACF to help identify useful lags.

The retail sales example has one seasonality with period 12 so let's use STL. If the time series has multiple seasonal periods (e.g., daily and weekly) then use MSTL to help detrend and deseasonalise the time series. We will show an example with multiple seasonalities in notebook 8 of this section.

```python
# break the time series into trend, seasonality, and residual components
res = STL(df_['y']).fit()

# plot the full decomposition
plt.rc('figure', figsize=(10, 10))
plt.rc('font', size=5)
res.plot();
```

```python
# plot ACF & PACF
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# ACF
plot_acf(x=res.resid,
        lags=36,
        ax=ax[0], 
        alpha=0.05,
        auto_ylims=True)

ax[0].set_title('Autocorrelation of residuals')
ax[0].set_ylabel('Autocorrelation')
ax[0].set_xlabel('Lag')

# PACF
plot_pacf(x=res.resid,
         lags=36,
         ax=ax[1],
         alpha=0.05,
         method='ywm',
         auto_ylims=True)

ax[1].set_title('Partial autocorrelation of residuals')
ax[1].set_ylabel('Partial autocorrelation')
ax[1].set_xlabel('Lag')

plt.tight_layout()
```


We can see that there aren't any large significant lags. There appears to be some significant lags at multiples of 12 which suggests that some part of the seasonal component is still in the residuals and was not perfectly extracted by STL. Practically speaking from looking at this plot there wouldn't be an additional lag beyond 1 or 2 that we would want to add for feature engineering purposes.

There we have it! We have shown how we can use the PACF for different types of time series to understand which lags could be useful as features. We will show how to use the PACF for a more challenging real world example in notebook 8.
